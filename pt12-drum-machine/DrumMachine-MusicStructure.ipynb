# More sophisticated beats

Music structure: verse, chorus

Examples at: http://beatsdrummachine.com/tutorial

So far we have done this:

```
Song:
  Tempo: 120
  Flow:
    - Verse:  x2

Verse:
  - bass.wav:       X...X...X...X...
  - snare.wav:      ....X.......X...
  - hh_closed.wav:  X.X.X.X.XX.XXXXX
```

We can have multiple pieces one after the other:

```
Song:
  Tempo: 120
  Flow:
    - Verse:  x4
    - Chorus: x4

Verse:
  - bass.wav:         X...X...X...X...
  - snare.wav:        ....X.......X...
  - hh_closed.wav:    X.X.X.X.XX.XXXXX

Chorus:
  - bass.wav:         XXXXXXXXXXXXX...
  - snare.wav:        ....X.......X...
  - hh_closed.wav:    XXXXXXXXXXXXX...
  - conga_low.wav:    X.....X.X..X....
  - conga_high.wav:   ....X....X......
  - cowbell_high.wav: ................

```

````python
# build the verse structure here:

bpm=120
beatv1 = makeBeat("beats/bass.wav", "X...X...X...X...", bpm)
verse = beatv1
# ... add beat 2 and 3
verse4 = repeatWave(verse, 4)
````

````python
# build the chorus structure here:
beatc1 = makeBeat("beats/bass.wav", "XXXXXXXXXXXXX...", bpm)
chorus = beatc1

chorus4 = repeatWave(chorus, 4)
````
