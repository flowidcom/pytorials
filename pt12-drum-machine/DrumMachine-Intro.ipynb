```python
from fwave import readWave, playWave, makeBeat, writeWave, repeatWave, makeAudio
```

```python
wave = readWave("beats/bass.wav")
makeAudio(wave)
```

```python
bpm = 120
```

```python
beat = makeBeat("beats/bass.wav", "X...X...X...X...", bpm)
```

```python
makeAudio(beat)
```

```python
beat1 = makeBeat("beats/bass.wav", "X...X...X...X...", bpm)
beat2 = makeBeat("beats/snare.wav", "....X.......X...", bpm)
beat3 = makeBeat("beats/hh_closed.wav", "X.X.X.X.XX.XXXXX", bpm)
verse = beat1 + beat2 + beat3
verse2 = repeatWave(verse, 2)
makeAudio(verse2)
```
